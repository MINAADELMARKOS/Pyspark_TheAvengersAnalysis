# **Import JSON**

```python
if __name__ == "__main__":
  spark = SparkSession \
        .builder \
        .master("local[*]") \
        .appName("Hello,Mina") \
        .getOrCreate()
  df = spark.read.json("/content/drive/MyDrive/Python_Learn /file1.json")
  df.printSchema()
  df.show()
  df.select(df['message'],df['restaurants']).show()
  df.filter(df['results_found'] > 45 ).show()
  df.filter(df['results_shown'] < 20).show()
  spark.stop
```

# **Aggregates**

```python
if __name__ == "__main__":
  spark = SparkSession \
        .builder \
        .master("local[*]") \
        .appName("Hello,Mina") \
        .getOrCreate()
  df = spark.read.json("/content/drive/MyDrive/Python_Learn /file1.json")
  df.printSchema()
  df.show()
  users = df.select(col("message").alias("mg"), col("results_found").cast("float").alias("rf"))

  users.printSchema()
  users.select(avg("rf")).show()
  users.select(max("rf")).show()
  users.select(min("rf")).show()

  spark.stop
```

# **Load/Save Functions**

```python
if __name__ == "__main__":
  spark = SparkSession \
        .builder \
        .master("local[*]") \
        .appName("parquet_json") \
        .getOrCreate()
  df = spark.read.load("/content/drive/MyDrive/Python_Learn /file1.json",format = "json")
  df.printSchema()

  df.select("message","results_found").write.save("message_results_found.parquet",format="parquet")


  spark.stop
```

# **Specifying a Data Source**

```python
if __name__ == "__main__":
  spark = SparkSession \
        .builder \
        .master("local[*]") \
        .appName("Iris") \
        .getOrCreate()

  df = spark.read.option("header",True).csv("/content/drive/MyDrive/Python_Learn /superheroes_info.csv")
  df.printSchema()
  df.show()
  df.columns

  dfIrisi = df.select(col('Name').cast("Double"),
                      col('Identity').cast("Double"),
                      col('Status').cast("Double"),
                      col('Alignment').cast("Double")
  )

  dfIrisi.write.save("iris-json",format = "json")
  spark.stop
```
